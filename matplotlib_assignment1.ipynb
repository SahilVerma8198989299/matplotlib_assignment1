A scatter plot is a type of data visualization that is used to display the relationship between two continuous variables. Each point on the plot represents a single observation in the data, with the x-coordinate of the point representing one variable and the y-coordinate representing the other variable.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a 2x2 grid of subplots
plt.figure(figsize=(12, 8))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Sin')

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Cos')

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Tan')

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Exp')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

A bar plot is a type of data visualization that is used to compare categorical data or to show the distribution of data across categories. In a bar plot, the length of each bar represents the value of a particular category, making it easy to compare values between categories visually.

Bar plots are useful when you want to compare the values of different categories or when you want to show how a single variable is distributed across different categories. They are commonly used in fields such as economics, business, and social sciences to visualize categorical data.

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit, color='skyblue')
```

A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution of a dataset along a single axis, showing the median, quartiles, and outliers. The box represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the box represents the median. The "whiskers" extend from the box to the minimum and maximum values of the dataset, excluding outliers, which are plotted as individual points.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.title('Box Plot of Box 1 and Box 2')
plt.show()
```
